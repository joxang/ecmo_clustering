```python
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np

#Python Library that deals with Bayesan Networks (BNs)
import pyAgrum as bn_graphs
import pyAgrum as gum
import pyAgrum.lib.notebook as gnb

from pyAgrum.lib.bn2roc import showROC

import seaborn as sns
sns.set()
```

```python
ecmodf = pd.read_csv('ecmodf_analysis_2.csv', index_col=0)
```

```python
con = ['age', 'bmi', 'ed_v', 'v_vv', 'sofa', 'pfr', 'pco2', 'pplat', 'lymph',
       'nlrat', 'pct', 'ferritin', 'crp', 'fib', 'ddim', 'time', 'resp', 'na']
cat = ['sex', 'death', 'ethnic', 'rrt', 'pe', 'ptx']
```

```python
ecmodf_bucketised = ecmodf[cat]
```

```python
for i in con:
    bucketised = pd.qcut(ecmodf[i], 4, precision=3, duplicates='drop')
    ecmodf_bucketised = pd.concat([ecmodf_bucketised, bucketised], axis=1, join='outer')
    
```

```python
bucketisedfile = 'ecmodf_bucketised.csv'
ecmodf_bucketised.to_csv(bucketisedfile, index=False)
```

```python
# call a BN learner
learner = gum.BNLearner( bucketisedfile )
# specify the learning algorithm. For the purposes of this unit, one can use the Hill Climbing algorithm
learner.useGreedyHillClimbing()
# learn the data
bn = learner.learnBN()
bn
```

```python
inference = bn_graphs.LazyPropagation(bn)

gnb.showInference(bn,inference)
```
